```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
import re

# CSV file path - change this to your actual file path
csv_file = '/content/ch_Ered.csv'  # Replace with your actual CSV file path


df = pd.read_csv(csv_file)
print(f"Successfully loaded {csv_file}")

# Print column names for verification
print("Columns in the dataset:")
for col in df.columns:
    print(f"- {col}")

# Define the display name mapping
method_mapping = {
    'Molecular_Laplace_BO_Ered': 'LLLA',
    'ibnn-Ered': 'I-BNN',
    'dkl-Ered': 'Deep kernel learning',
    'gp-Ered': 'GP',
    'Molecular_VBLL_Ered': 'VBLL',
    'Molecular_BLOB_Ered_dense_precision_multirank': 'VBLL',
    'BLOB_EnsembleMolecular_BLOB_Ered': 'BLOB',
    'Molecular_BLOB_Ered_mlp': 'BLOB',
    'Ered_mlp': 'Small NN VBLL',
    'Molecular_DeepEnsemble_Ered': 'Deep Ensemble',
    'Molecular_DeepEnsemble_Ered_vbll': 'Deep Ensemble',
    'Molecular_DeepEnsemble_Ered_std': 'Deep Ensemble',
}

# Function to determine the display name for each column
def get_display_name(column_name):
    for key, value in method_mapping.items():
        if key in column_name:
            return value
    return column_name

# Print mapping for verification
print("\nMethod mappings:")
for col in df.columns:
    if col != 'Step':
        display_name = get_display_name(col)
        print(f"{col} -> {display_name}")

# Fill empty values in BLOB columns with previous values
blob_columns = [col for col in df.columns if 'BLOB' in col]
print(f"\nFilling missing values in {len(blob_columns)} BLOB columns using forward fill")
for col in blob_columns:
    # Check for null values before filling
    null_count = df[col].isnull().sum()
    if null_count > 0:
        print(f"- {col}: {null_count} missing values found")
    df[col] = df[col].fillna(method='ffill')

target_length = 150

# 确保所有列的长度都相同，并且达到目标长度
print(f"\n检查每列数据长度，目标长度为: {target_length}")

# 首先确保有一个完整的Step列，从0到target_length-1
if 'Step' not in df.columns:
    print("创建Step列，因为它不存在")
    df['Step'] = np.arange(0, len(df))

# 检查Step列是否完整，如果不完整则补全
max_step = df['Step'].max()
if max_step < target_length - 1:
    missing_steps = list(range(int(max_step) + 1, target_length))
    print(f"Step列不完整，补充步骤: {len(missing_steps)}个")

    # 创建新行，每个缺失的步骤一行
    new_rows = pd.DataFrame({'Step': missing_steps})
    for col in df.columns:
        if col != 'Step':
            new_rows[col] = np.nan  # 先用NaN填充，后面单独处理每列

    # 将新行添加到原始DataFrame
    df = pd.concat([df, new_rows], ignore_index=True)

    # 确保Step列是排序的
    df = df.sort_values('Step').reset_index(drop=True)

# 检查每列的有效值长度，如果不足则用最后一个有效值填充
for col in df.columns:
    if col == 'Step':
        continue

    # 检查该列的null值数量
    null_mask = df[col].isnull()
    null_count = null_mask.sum()

    if null_count > 0:
        print(f"列 '{col}' 有 {null_count} 个缺失值")

        # 找到最后一个非null值
        last_valid_idx = df[~null_mask][col].index.max() if sum(~null_mask) > 0 else None

        if last_valid_idx is not None:
            last_valid_value = df.loc[last_valid_idx, col]
            print(f"  用最后一个有效值 {last_valid_value} 填充剩余位置")

            # 将最后一个有效值填充到所有后续的null位置
            for idx in df[null_mask].index:
                if idx > last_valid_idx:
                    df.loc[idx, col] = last_valid_value
        else:
            print(f"  列 '{col}' 没有有效值，用0填充")
            df[col] = df[col].fillna(0)

# 验证数据现在是否达到目标长度
print(f"Final data shape: {df.shape}, with Step range: {df['Step'].min()} to {df['Step'].max()}")

# Group columns by method for calculations
method_columns = {}
for col in df.columns:
    if col == 'Step':
        continue
    display_name = get_display_name(col)
    if display_name not in method_columns:
        method_columns[display_name] = []
    method_columns[display_name].append(col)

# Print the number of variants for each method
print("\nNumber of variants for each method:")
for method, cols in method_columns.items():
    print(f"{method}: {len(cols)} variants")
    if len(cols) > 0:
        print(f"  - {', '.join(cols[:3])}" + ("..." if len(cols) > 3 else ""))

# Set up the figure and axis with a modern style
plt.style.use('seaborn-v0_8-whitegrid')
plt.figure(figsize=(14, 10))

# Define colors for each method (optimized for visibility)
colors = {
    'BLOB': '#1f77b4',     # blue
    'LLLA': '#ff7f0e',     # orange
    'I-BNN': '#2ca02c',    # green
    'Deep kernel learning': '#d62728',  # red
    'GP': '#9467bd',       # purple
    'VBLL': '#8c564b',     # brown
    'Small NN VBLL': '#e377c2',  # pink
    'Deep Ensemble': '#17becf'   # cyan
}

# Define markers for each method
markers = {
    'BLOB': 'o',
    'LLLA': 's',
    'I-BNN': '^',
    'Deep kernel learning': 'D',
    'GP': 'p',
    'VBLL': '*',
    'Small NN VBLL': 'x',
    'Deep Ensemble': 'P'  # plus (filled)
}

# Plot each method with mean as solid line and variance as shaded area
for method, cols in method_columns.items():
    if len(cols) > 0:
        # Create a DataFrame with just these columns for easier manipulation
        method_df = df[['Step'] + cols].copy()

        # Calculate mean and std for this method across all variants
        method_df['mean'] = method_df[cols].mean(axis=1)
        method_df['std'] = method_df[cols].std(axis=1)

        # Some methods might have only one variant, in which case std will be NaN or 0
        if len(cols) == 1 or method_df['std'].isnull().all():
            method_df['std'] = method_df['mean'] * 0.02  # Set a small artificial variance for visibility
            print(f"Note: {method} has only one variant, using artificial variance for visualization")

        # Plot shaded area for variance (before the mean line so it's behind)
        plt.fill_between(
            method_df['Step'],
            method_df['mean'] - 0.1*method_df['std'],
            method_df['mean'] + 0.1*method_df['std'],
            color=colors.get(method, 'black'),
            alpha=0.2,
            label=f"{method} ±σ"
        )

        # Plot mean line
        plt.plot(
            method_df['Step'],
            method_df['mean'],
            label=f"{method} mean",
            color=colors.get(method, 'black'),
            linewidth=3,
            marker=markers.get(method, 'o'),
            markevery=max(1, len(df) // 20),
            markersize=8
        )

# Add labels and legend
plt.xlabel('Step', fontsize=16, fontweight='bold')
plt.ylabel('Redox Potential', fontsize=16, fontweight='bold')
plt.title('Redoxmer (minimize)', fontsize=20, fontweight='bold')
plt.grid(True, linestyle='--', alpha=0.7)

# Create a custom legend that shows only one entry per method
handles, labels = plt.gca().get_legend_handles_labels()
unique_methods = list(method_columns.keys())

# This will store our final handles and labels for the legend
legend_handles = []
legend_labels = []

# Process each unique method and keep only one entry for each
for method in unique_methods:
    # Find the first occurrence of this method in the labels
    for i, label in enumerate(labels):
        if method in label:
            legend_handles.append(handles[i])
            legend_labels.append(method)
            break

# Create the legend with our custom handles and labels
plt.legend(
    legend_handles,
    legend_labels,
    fontsize=12,
    loc='best',
    framealpha=0.9,
    edgecolor='gray',
    fancybox=True,
    shadow=True
)

# Adjust the plot aesthetics
plt.xticks(np.arange(0, 150, step=20), fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()

plt.xlim(0, 150)

# Set y-axis limits if needed
plt.ylim(1.54, 2.0)  # Corrected from plt.set_ylim to plt.ylim


# Add a border around the plot
ax = plt.gca()
for spine in ax.spines.values():
    spine.set_visible(True)
    spine.set_linewidth(1.5)
    spine.set_color('gray')

# Save the figure with a timestamp to avoid overwriting
import datetime
timestamp = datetime.datetime.now().strftime("%Y%m%d_%H%M%S")
output_file = f'model_comparison_mean_var_{timestamp}_final_8_en.png'
plt.savefig(output_file, dpi=300, bbox_inches='tight')
print(f"\nFigure saved as {output_file}")

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
import re

# CSV file path - change this to your actual file path
csv_file = '/content/RE-TRIGGER-FINE_TUNING.csv'  # Replace with your actual CSV file path


df = pd.read_csv(csv_file)
print(f"Successfully loaded {csv_file}")

# Print column names for verification
print("Columns in the dataset:")
for col in df.columns:
    print(f"- {col}")

# Define the display name mapping
method_mapping = {
    'Molecular_BLOB_Ered_dense_precision_multirank_fine-tuning': 'Fine-tuning',
    'ibnn-Ered': 'I-BNN',
    'dkl-Ered': 'Deep kernel learning',
    'gp-Ered': 'GP',
    'Ered_mlp': 'Small NN VBLL',
    'Molecular_BLOB_Ered_dense_precision_multirank_ml_trigger': 'Event_trigger Update',
    'Molecular_BLOB_Ered_dense_precision_multirank_recursive': 'Recursive Update',

}

# Function to determine the display name for each column
def get_display_name(column_name):
    for key, value in method_mapping.items():
        if key in column_name:
            return value
    return column_name

# Print mapping for verification
print("\nMethod mappings:")
for col in df.columns:
    if col != 'Step':
        display_name = get_display_name(col)
        print(f"{col} -> {display_name}")

# Fill empty values in BLOB columns with previous values
blob_columns = [col for col in df.columns if 'BLOB' in col]
print(f"\nFilling missing values in {len(blob_columns)} BLOB columns using forward fill")
for col in blob_columns:
    # Check for null values before filling
    null_count = df[col].isnull().sum()
    if null_count > 0:
        print(f"- {col}: {null_count} missing values found")
    df[col] = df[col].fillna(method='ffill')

target_length = 150

# 确保所有列的长度都相同，并且达到目标长度
print(f"\n检查每列数据长度，目标长度为: {target_length}")

# 首先确保有一个完整的Step列，从0到target_length-1
if 'Step' not in df.columns:
    print("创建Step列，因为它不存在")
    df['Step'] = np.arange(0, len(df))

# 检查Step列是否完整，如果不完整则补全
max_step = df['Step'].max()
if max_step < target_length - 1:
    missing_steps = list(range(int(max_step) + 1, target_length))
    print(f"Step列不完整，补充步骤: {len(missing_steps)}个")

    # 创建新行，每个缺失的步骤一行
    new_rows = pd.DataFrame({'Step': missing_steps})
    for col in df.columns:
        if col != 'Step':
            new_rows[col] = np.nan  # 先用NaN填充，后面单独处理每列

    # 将新行添加到原始DataFrame
    df = pd.concat([df, new_rows], ignore_index=True)

    # 确保Step列是排序的
    df = df.sort_values('Step').reset_index(drop=True)

# 检查每列的有效值长度，如果不足则用最后一个有效值填充
for col in df.columns:
    if col == 'Step':
        continue

    # 检查该列的null值数量
    null_mask = df[col].isnull()
    null_count = null_mask.sum()

    if null_count > 0:
        print(f"列 '{col}' 有 {null_count} 个缺失值")

        # 找到最后一个非null值
        last_valid_idx = df[~null_mask][col].index.max() if sum(~null_mask) > 0 else None

        if last_valid_idx is not None:
            last_valid_value = df.loc[last_valid_idx, col]
            print(f"  用最后一个有效值 {last_valid_value} 填充剩余位置")

            # 将最后一个有效值填充到所有后续的null位置
            for idx in df[null_mask].index:
                if idx > last_valid_idx:
                    df.loc[idx, col] = last_valid_value
        else:
            print(f"  列 '{col}' 没有有效值，用0填充")
            df[col] = df[col].fillna(0)

# 验证数据现在是否达到目标长度
print(f"Final data shape: {df.shape}, with Step range: {df['Step'].min()} to {df['Step'].max()}")

# Group columns by method for calculations
method_columns = {}
for col in df.columns:
    if col == 'Step':
        continue
    display_name = get_display_name(col)
    if display_name not in method_columns:
        method_columns[display_name] = []
    method_columns[display_name].append(col)

# Print the number of variants for each method
print("\nNumber of variants for each method:")
for method, cols in method_columns.items():
    print(f"{method}: {len(cols)} variants")
    if len(cols) > 0:
        print(f"  - {', '.join(cols[:3])}" + ("..." if len(cols) > 3 else ""))

# Set up the figure and axis with a modern style
plt.style.use('seaborn-v0_8-whitegrid')
plt.figure(figsize=(14, 10))

# Define colors for each method (optimized for visibility)
colors = {
    'BLOB': '#1f77b4',     # blue
    'Fine-tuning': '#ff7f0e',     # orange
    'I-BNN': '#2ca02c',    # green
    'Deep kernel learning': '#d62728',  # red
    'GP': '#9467bd',       # purple
    'Event_trigger Update': '#8c564b',     # brown
    'Small NN VBLL': '#e377c2',  # pink
    'Recursive Update': '#17becf'   # cyan
}

# Define markers for each method
markers = {
    'BLOB': 'o',
    'Fine-tuning': 's',
    'I-BNN': '^',
    'Deep kernel learning': 'D',
    'GP': 'p',
    'Event_trigger Update': '*',
    'Small NN VBLL': 'x',
    'Recursive Update': 'P'  # plus (filled)
}

# Plot each method with mean as solid line and variance as shaded area
for method, cols in method_columns.items():
    if len(cols) > 0:
        # Create a DataFrame with just these columns for easier manipulation
        method_df = df[['Step'] + cols].copy()

        # Calculate mean and std for this method across all variants
        method_df['mean'] = method_df[cols].mean(axis=1)
        method_df['std'] = method_df[cols].std(axis=1)

        # Some methods might have only one variant, in which case std will be NaN or 0
        if len(cols) == 1 or method_df['std'].isnull().all():
            method_df['std'] = method_df['mean'] * 0.02  # Set a small artificial variance for visibility
            print(f"Note: {method} has only one variant, using artificial variance for visualization")

        # Plot shaded area for variance (before the mean line so it's behind)
        plt.fill_between(
            method_df['Step'],
            method_df['mean'] - 0.1*method_df['std'],
            method_df['mean'] + 0.1*method_df['std'],
            color=colors.get(method, 'black'),
            alpha=0.2,
            label=f"{method} ±σ"
        )

        # Plot mean line
        plt.plot(
            method_df['Step'],
            method_df['mean'],
            label=f"{method} mean",
            color=colors.get(method, 'black'),
            linewidth=3,
            marker=markers.get(method, 'o'),
            markevery=max(1, len(df) // 20),
            markersize=8
        )

# Add labels and legend
plt.xlabel('Step', fontsize=16, fontweight='bold')
plt.ylabel('Redox Potential', fontsize=16, fontweight='bold')
plt.title('Redoxmer (minimize)', fontsize=20, fontweight='bold')
plt.grid(True, linestyle='--', alpha=0.7)

# Create a custom legend that shows only one entry per method
handles, labels = plt.gca().get_legend_handles_labels()
unique_methods = list(method_columns.keys())

# This will store our final handles and labels for the legend
legend_handles = []
legend_labels = []

# Process each unique method and keep only one entry for each
for method in unique_methods:
    # Find the first occurrence of this method in the labels
    for i, label in enumerate(labels):
        if method in label:
            legend_handles.append(handles[i])
            legend_labels.append(method)
            break

# Create the legend with our custom handles and labels
plt.legend(
    legend_handles,
    legend_labels,
    fontsize=12,
    loc='best',
    framealpha=0.9,
    edgecolor='gray',
    fancybox=True,
    shadow=True
)

# Adjust the plot aesthetics
plt.xticks(np.arange(0, 150, step=20), fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()

plt.xlim(0, 150)

# Set y-axis limits if needed
plt.ylim(1.54, 2.0)  # Corrected from plt.set_ylim to plt.ylim


# Add a border around the plot
ax = plt.gca()
for spine in ax.spines.values():
    spine.set_visible(True)
    spine.set_linewidth(1.5)
    spine.set_color('gray')

# Save the figure with a timestamp to avoid overwriting
import datetime
timestamp = datetime.datetime.now().strftime("%Y%m%d_%H%M%S")
output_file = f'model_comparison_mean_var_{timestamp}_final_RE-TRIGGER-FINE_TUNING.png'
plt.savefig(output_file, dpi=300, bbox_inches='tight')
print(f"\nFigure saved as {output_file}")

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
import re
import datetime

# CSV file path - change this to your actual file path
csv_file = '/content/PestControl.csv'  # Replace with your actual CSV file path

# Target length for all data series (you can adjust this value)
target_length = 220  # 设置目标长度，所有数据将填充到这个长度

# Check if file exists
if not os.path.exists(csv_file):
    print(f"File {csv_file} not found. Please check the file path.")
    # Create a sample DataFrame for testing
    steps = np.arange(0, 100, 5)
    data = {
        'Step': steps,
        'Molecular_BLOB_Ered_mlp_20250508_232820 - best_value': np.random.rand(len(steps)) * 0.5 - 1.0,
        'Molecular_Laplace_BO_Ered_20250507_212946 - best_value': np.random.rand(len(steps)) * 0.5 - 1.2,
        'ibnn-Ered - best_value': np.random.rand(len(steps)) * 0.4 - 1.3,
        'dkl-Ered - best_value': np.random.rand(len(steps)) * 0.6 - 1.1,
        'gp-Ered - best_value': np.random.rand(len(steps)) * 0.3 - 1.4,
        'Molecular_VBLL_Ered_dense_precision_multirank_20250507_014507 - best_value': np.random.rand(len(steps)) * 0.7 - 1.0,
        'BLOB_EnsembleMolecular_BLOB_Ered_20250508_015139 - best_y': np.random.rand(len(steps)) * 0.5 - 1.2,
        'BLOB_EnsembleMolecular_BLOB_Ered_20250508_122413 - best_value': np.random.rand(len(steps)) * 0.4 - 1.1
    }
    df = pd.DataFrame(data)
    print("Using sample data for testing")
else:
    # Read the CSV file
    df = pd.read_csv(csv_file)
    print(f"Successfully loaded {csv_file}")

# Print column names for verification
print("Columns in the dataset:")
for col in df.columns:
    print(f"- {col}")

# Define the display name mapping
method_mapping = {
    'CSV_BLOB_pest_control': 'LLLA',
    'ibnn-Gsol': 'I-BNN',
    'dkl-Gsol': 'Deep kernel learning',
    'gp-Gsol': 'GP',
    'CSV_VBLL_pest_control': 'VBLL',
    #'CSV_VBLL_pest_control_dense_precision_multirank': 'VBLL',
    'BLOB_EnsembleEnsemble_BLOB_pest_control': 'BLOB',
    'pest_control_mlp': 'Small NN VBLL',
    'DeepEnsemble_pest_control': 'Deep Ensemble',
    'DeepEnsemble_pest_control_vbll': 'Deep Ensemble',
    'DeepEnsemble_pest_control_std': 'Deep Ensemble',
}

# Function to determine the display name for each column
def get_display_name(column_name):
    for key, value in method_mapping.items():
        if key in column_name:
            return value
    return column_name

# Print mapping for verification
print("\nMethod mappings:")
for col in df.columns:
    if col != 'Step':
        display_name = get_display_name(col)
        print(f"{col} -> {display_name}")

# Fill empty values in BLOB columns with previous values
blob_columns = [col for col in df.columns if 'BLOB' in col]
print(f"\nFilling missing values in {len(blob_columns)} BLOB columns using forward fill")
for col in blob_columns:
    # Check for null values before filling
    null_count = df[col].isnull().sum()
    if null_count > 0:
        print(f"- {col}: {null_count} missing values found")
    df[col] = df[col].fillna(method='ffill')

# 检查Step列是否已经存在，如果不存在则创建
if 'Step' not in df.columns:
    print("Creating Step column since it doesn't exist")
    df['Step'] = np.arange(0, len(df))

# 确保所有列的长度都相同，并且达到目标长度
print(f"\n检查每列数据长度，目标长度为: {target_length}")

# 首先确保有一个完整的Step列，从0到target_length-1
if 'Step' not in df.columns:
    print("创建Step列，因为它不存在")
    df['Step'] = np.arange(0, len(df))

# 检查Step列是否完整，如果不完整则补全
max_step = df['Step'].max()
if max_step < target_length - 1:
    missing_steps = list(range(int(max_step) + 1, target_length))
    print(f"Step列不完整，补充步骤: {len(missing_steps)}个")

    # 创建新行，每个缺失的步骤一行
    new_rows = pd.DataFrame({'Step': missing_steps})
    for col in df.columns:
        if col != 'Step':
            new_rows[col] = np.nan  # 先用NaN填充，后面单独处理每列

    # 将新行添加到原始DataFrame
    df = pd.concat([df, new_rows], ignore_index=True)

    # 确保Step列是排序的
    df = df.sort_values('Step').reset_index(drop=True)

# 检查每列的有效值长度，如果不足则用最后一个有效值填充
for col in df.columns:
    if col == 'Step':
        continue

    # 检查该列的null值数量
    null_mask = df[col].isnull()
    null_count = null_mask.sum()

    if null_count > 0:
        print(f"列 '{col}' 有 {null_count} 个缺失值")

        # 找到最后一个非null值
        last_valid_idx = df[~null_mask][col].index.max() if sum(~null_mask) > 0 else None

        if last_valid_idx is not None:
            last_valid_value = df.loc[last_valid_idx, col]
            print(f"  用最后一个有效值 {last_valid_value} 填充剩余位置")

            # 将最后一个有效值填充到所有后续的null位置
            for idx in df[null_mask].index:
                if idx > last_valid_idx:
                    df.loc[idx, col] = last_valid_value
        else:
            print(f"  列 '{col}' 没有有效值，用0填充")
            df[col] = df[col].fillna(0)

# 验证数据现在是否达到目标长度
print(f"Final data shape: {df.shape}, with Step range: {df['Step'].min()} to {df['Step'].max()}")

# Group columns by method for calculations
method_columns = {}
for col in df.columns:
    if col == 'Step':
        continue
    display_name = get_display_name(col)
    if display_name not in method_columns:
        method_columns[display_name] = []
    method_columns[display_name].append(col)

# Print the number of variants for each method
print("\nNumber of variants for each method:")
for method, cols in method_columns.items():
    print(f"{method}: {len(cols)} variants")
    if len(cols) > 0:
        print(f"  - {', '.join(cols[:3])}" + ("..." if len(cols) > 3 else ""))

# Set up the figure and axis with a modern style
plt.style.use('seaborn-v0_8-whitegrid')
plt.figure(figsize=(14, 10))

# Define colors for each method (optimized for visibility)
colors = {
    'BLOB': '#1f77b4',     # blue
    'LLLA': '#ff7f0e',     # orange
    'I-BNN': '#2ca02c',    # green
    'Deep kernel learning': '#d62728',  # red
    'GP': '#9467bd',       # purple
    'VBLL': '#8c564b',     # brown
    'Small NN VBLL': '#e377c2',  # pink
    'Deep Ensemble': '#17becf'   # cyan
}

# Define markers for each method
markers = {
    'BLOB': 'o',
    'LLLA': 's',
    'I-BNN': '^',
    'Deep kernel learning': 'D',
    'GP': 'p',
    'VBLL': '*',
    'Small NN VBLL': 'x',
    'Deep Ensemble': 'P'  # plus (filled)
}

# 存储每个方法的平均线句柄用于图例
method_line_handles = {}
method_line_labels = {}

# Plot each method with mean as solid line and variance as shaded area
for method, cols in method_columns.items():
    if len(cols) > 0:
        # Create a DataFrame with just these columns for easier manipulation
        method_df = df[['Step'] + cols].copy()

        # Calculate mean and std for this method across all variants
        method_df['mean'] = method_df[cols].mean(axis=1)
        method_df['std'] = method_df[cols].std(axis=1)

        # Some methods might have only one variant, in which case std will be NaN or 0
        if len(cols) == 1 or method_df['std'].isnull().all():
            method_df['std'] = method_df['mean'] * 0.02  # Set a small artificial variance for visibility
            print(f"Note: {method} has only one variant, using artificial variance for visualization")

        # Plot shaded area for variance (before the mean line so it's behind)
        plt.fill_between(
            method_df['Step'],
            method_df['mean'] - 0.1 * method_df['std'],
            method_df['mean'] + 0.1 * method_df['std'],
            color=colors.get(method, 'black'),
            alpha=0.2
        )

        # Plot mean line and save its handle for the legend
        line, = plt.plot(
            method_df['Step'],
            method_df['mean'],
            color=colors.get(method, 'black'),
            linewidth=3,
            marker=markers.get(method, 'o'),
            markevery=max(1, len(df) // 20),
            markersize=8
        )

        # 保存这个方法的线条句柄
        method_line_handles[method] = line
        method_line_labels[method] = method

# 现在我们直接使用存储的线条句柄创建图例
plt.legend(
    list(method_line_handles.values()),
    list(method_line_labels.values()),
    fontsize=12,
    loc='best',
    framealpha=0.9,
    edgecolor='gray',
    fancybox=True,
    shadow=True
)

# Add labels
plt.xlabel('Step', fontsize=16, fontweight='bold')
plt.ylabel('Cost', fontsize=16, fontweight='bold')
plt.title('Pest Control', fontsize=20, fontweight='bold')
plt.grid(True, linestyle='--', alpha=0.7)

# Adjust the plot aesthetics
plt.xticks(np.arange(0, target_length + 20, step=20), fontsize=12)  # 更新X轴刻度以适应新的目标长度
plt.yticks(fontsize=12)
plt.tight_layout()

plt.xlim(0, target_length)  # 更新X轴显示范围
plt.ylim(12.5, 13)

# Add a border around the plot
ax = plt.gca()
for spine in ax.spines.values():
    spine.set_visible(True)
    spine.set_linewidth(1.5)
    spine.set_color('gray')

# Save the figure with a timestamp to avoid overwriting
timestamp = datetime.datetime.now().strftime("%Y%m%d_%H%M%S")
output_file = f'model_comparison_mean_var_{timestamp}_final.png'
plt.savefig(output_file, dpi=300, bbox_inches='tight')
print(f"\nFigure saved as {output_file}")

# Show the plot
plt.show()
```
